```python
import pandas as pd

# Get existing show_id values from SQL Server
query = "SELECT show_id FROM netflix_raw"
existing_ids = pd.read_sql(query, conn)

# Remove existing show_id from df
df = df[~df['show_id'].isin(existing_ids['show_id'])]

# Insert cleaned data
df.to_sql('netflix_raw', con=conn, index=False, if_exists='append')
print("✅ Data Inserted Successfully (No Duplicates)")
```
